### Приведение названий столбцов к нижнему регистру

### Reducing column names to lowercase

```python
nested = list(data)
# Подфункция перебирания списка
def apply_nested(func, lst, isatom=lambda item: not isinstance(item, list)):
    for i, item in enumerate(lst):
        if isatom(item):
            lst[i] = func(item)
        else:
            apply_nested(func, item, isatom)
            
apply_nested(lambda atom: atom.lower(), nested)
data.columns = nested
data.info()
```

Выполнена замена названий столбцов на нижний регистр.

Column names were replaced with lower case.

### Column 'name'

#### Converting game names to lowercase

```python
data['name'] = data['name'].str.lower()
```

#### Очистка от пропусков

#### Clearing of missing values

```python
display(data[data['name'].isna()])
data = data.loc[~data['name'].isna()]
print(data.isna().sum())
print()
print('Количество уникальных названий игр:', len(data['name'].unique()))
```

Найдены две записи с пустыми названиями в столбце 'name' и полностью удалены, т.к. нет названия - нет игры.

Выявлено, что есть игры с одинаковыми названиями. Возможно, они продавались в разные годы и/или на разных платформах.

Two entries with empty names were found in the 'name' column and completely deleted, because there is no name - there is no game.

It is revealed that there are games with the same names. They may have been sold in different years and/or on different platforms.

#### Filling missing values by the name of the game

```python
print(data['critic_score'].describe())
# Подфункция заполнения пропусков по медиане
def critic_score_finder(row):
    name         = row['name']
    critic_score = row['critic_score']
    # Срез для текущей игры
    if np.isnan(critic_score):
        array = data[data['name'] == name]['critic_score']
        if ~np.all(np.isnan(array)):
            return np.nanmedian(array)
        else:
            return critic_score
    else:
        return critic_score
    
data['critic_score'] = data.apply(critic_score_finder, axis = 1)
print()
print('Осталось нулевых значений: ', data['critic_score'].isna().sum())
```

```python
def popular_platforms(region_sales):
    # Популярные (ТОП-5) платформы
    # получаем сводную таблицу
    print('ТОП-5 популярных платформ в регионе:')
    platform_p_sales_N_yrs = pd.pivot_table(data_N_yrs, index = 'platform', values = region_sales, aggfunc = sum)
    region_total_sales = platform_p_sales_N_yrs[region_sales].sum()
    platform_p_sales_N_yrs['part_in_%'] = platform_p_sales_N_yrs[region_sales]/region_total_sales*100
    # сортируем по убыванию прибыли
    platform_p_sales_N_yrs = platform_p_sales_N_yrs.sort_values(by = region_sales, ascending = False)
    # сбрасываем индексы и оставляем ТОП-5 платформ
    platform_p_sales_N_yrs.reset_index(inplace = True)
    platform_p_sales_N_yrs.drop(index = list(range(5, len(platform_p_sales_N_yrs))), inplace = True)
    region_top_5_sales = platform_p_sales_N_yrs[region_sales].sum()
    # Добавляем строку others
    new_row = {'platform' : 'others',
              region_sales : region_total_sales-region_top_5_sales,
              'part_in_%' : (region_total_sales-region_top_5_sales)/region_total_sales*100}
    platform_p_sales_N_yrs = platform_p_sales_N_yrs.append(new_row, ignore_index = True)
    display(platform_p_sales_N_yrs)
    # Строим график по прибылям
    platform_p_sales_N_yrs.plot(x = 'platform', y = region_sales, kind = 'bar').grid(axis = 'y')
    plt.title('График продаж по Топ-5 платформам за последние {} лет'.format(N_yrs))
    plt.show()
    # Строим круговую диаграмму
    platform_p_sales_N_yrs.plot.pie(y = region_sales,
                                    autopct = '%1.0f%%',
                                    labels = platform_p_sales_N_yrs['platform'],
                                   figsize = (15, 10))
    plt.title('Pie-диаграмма долей продаж по Топ-5 платформам за последние {} лет'.format(N_yrs))
    plt.show()
    # =======================================================================================================
    print('')
    print('')
    print('Рейтинг ESRB и его влияние в регионе:')
    # Влияние рейтинга ESRB
    rating_p_sales_N_yrs = pd.pivot_table(data_N_yrs, index = 'rating', values = region_sales, aggfunc = sum)
    # сортируем по убыванию прибыли
    rating_p_sales_N_yrs = rating_p_sales_N_yrs.sort_values(by = region_sales, ascending = False)
    display(rating_p_sales_N_yrs)
    # Строим график по прибылям
    rating_p_sales_N_yrs.plot(kind = 'bar').grid(axis = 'y')
    plt.title('График продаж по рейтингу за последние {} лет'.format(N_yrs))
    plt.show()
    # Строим круговую диаграмму
    rating_p_sales_N_yrs.plot.pie(y = region_sales,
                                    autopct = '%1.0f%%',
                                    figsize = (15, 10))
    plt.title('Pie-диаграмма рейтинга ESRB за последние {} лет'.format(N_yrs))
    plt.show()
    
    
popular_platforms('na_sales')
```

##  Проверка гипотез

## Hypothesis testing

### Средние пользовательские рейтинги 'xone' и 'pc' одинаковы

### The average user ratings of 'xone' and ' pc ' are the same

```python
series_1 = data_N_yrs[data_N_yrs['platform'] == 'xone']['user_score']
series_1.dropna(inplace = True)
print('Средний user_score по xone:', series_1.mean())
series_2 = data_N_yrs[data_N_yrs['platform'] == 'pc']['user_score']
series_2.dropna(inplace = True)
print('Средний user_score по pc:', series_2.mean())
alpha = 0.05
results_1 = st.ttest_ind(series_1, series_2)

print()
print('p-значение:', results_1.pvalue)

if results_1.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Сформулированы гипотезы:
 - H0 - средние значения множеств равны
 - H1 - средние значения множеств не равны
 - Уровень значимости = 5%, как компромисс между значимостью и мощностью множеств. Можно было и взять более строгий уровень (например, 1%).

В результате:


p-значение ~24% и это очень высокое значение.

Не получилось отвергнуть H0 (средние значения множеств равны), даже если брать менее строгие значения alpha.


Hypotheses are formulated:
 - H0 - the average values of the sets are equal to
 - H1 - the average values of the sets are not equal
 - Significance level = 5%, as a compromise between the significance and the power of the sets. It was possible to take a more strict level (for example, 1%).

As a result,:


the p-value is ~24% and this is a very high value.

It was not possible to reject H0 (the average values of the sets are equal), even if we take less strict alpha values.
