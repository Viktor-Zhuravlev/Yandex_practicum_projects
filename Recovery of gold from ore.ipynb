### Расчет эффективности обогащения `recovery` на обучающей выборке

### Calculation of the efficiency of the `recovery` enrichment on the training sample

```python
# Подфункция для расчета эффективности обогащения
def recovery_calc(C, F, T):
    recovery = (C*(F-T))/(F*(C-T))*100
    return recovery

recovery = recovery_calc(data_train['rougher.output.concentrate_au'],
                        data_train['rougher.input.feed_au'],
                        data_train['rougher.output.tail_au'])
```

```python
fig = plt.figure()

ax = fig.add_subplot(111, label = "1")
ax2 = fig.add_subplot(111, label = "2", frame_on = False)

ax.plot(data_train['rougher.output.recovery'], color = "C0")

ax2.plot(recovery, color = "C3")

plt.title('Графики эффективности обогащения')
plt.show()
```

#### Расчет MAE для эффективности обогащения

#### Calculation of MAE for enrichment efficiency

```python
print(mean_absolute_error(data_train['rougher.output.recovery'], recovery))
```

Проведен расчет эффективности обогащения и вычислена MAE для расчетной и имеющейся в массиве. Порядок ошибки 10**(-15), свидетельствует о верном расчете.

The calculation of the enrichment efficiency is carried out and the MAE is calculated for the calculated and available in the array. The order of the error is 10**(-15), indicates a correct calculation.

### Вычисление sMAPE и total_sMAPE

### Calculation sMAPE и total_sMAPE

```python
def sMAPE_calc(target, prediction):
    A = target-prediction
    B = (np.abs(target)+np.abs(prediction))/2
    C = np.abs(A)/B
    return np.mean(C)*100

def total_sMAPE_calc(rougher, final):
    return 0.25*rougher+0.75*final
```

#### Linear regression

```python
def best_params_LinearRegression(features, target):
    # -------------------------------------------------------
    # Масштабирование
    scaler_features = StandardScaler()
    scaler_features.fit(features)
    features_mapped = scaler_features.transform(features)
    # -------------------------------------------------------
    # Тип модели и её параметры
    model = LinearRegression()
    parameters = {'normalize': [True, False]}
    # -------------------------------------------------------
    sMAPE_s = make_scorer(sMAPE_calc, greater_is_better = False)
    grid = GridSearchCV(model, parameters, scoring = sMAPE_s, cv = 5)
    grid.fit(features_mapped, target)
    print('Лучшие параметры:', grid.best_params_)
    # -------------------------------------------------------
    # Обучение модели
    best_model = LinearRegression(normalize = grid.best_params_['normalize'])
    best_model.fit(features_mapped, target)
    # -------------------------------------------------------
    # Моделирование и метрики
    predictions = best_model.predict(features_mapped)
    sMAPE = sMAPE_calc(target, predictions)
    print('sMAPE модели регрессии на обучающей выборке:', sMAPE)
    print("Коэффициент детерминации на обучающей выборке:", r2_score(target, predictions))
    # -------------------------------------------------------
    # Строим график
    plt.scatter(target, predictions, color='black')
    plt.xlabel('Target')
    plt.ylabel('Predictions')
    plt.title('График Target-Predictions на обучающей выборке')
    plt.show()
    return best_model, scaler_features
```

```python
def best_model_simulation(model, scaler_features, features, target):
    # -------------------------------------------------------
    features_mapped = scaler_features.transform(features)
    predictions = model.predict(features_mapped)
    # -------------------------------------------------------
    # Моделирование и метрики
    predictions = model.predict(features_mapped)
    sMAPE = sMAPE_calc(target, predictions)
    print('sMAPE модели на тестовой выборке:', sMAPE)
    print("Коэффициент детерминации на тестовой выборке:", r2_score(target, predictions))
    # -------------------------------------------------------
    # Строим график
    plt.scatter(target, predictions, color = 'black')
    plt.xlabel('Target')
    plt.ylabel('Predictions')
    plt.title('График Target-Predictions на тестовой выборке')
    plt.show()
    return sMAPE
```

#### Random forest

```python
def best_params_RandomForestRegression(features, target):
    # -------------------------------------------------------
    # Масштабирование
    scaler_features = StandardScaler()
    scaler_features.fit(features)
    features_mapped = scaler_features.transform(features)
    # -------------------------------------------------------
    # Тип модели и её параметры
    model = RandomForestRegressor()
    parameters = {'n_estimators': range(10, 100, 10),
                 'max_depth': range(1, 10)}
    # -------------------------------------------------------
    sMAPE_s = make_scorer(sMAPE_calc, greater_is_better = False)
    grid = GridSearchCV(model, parameters, scoring = sMAPE_s, cv = 5, n_jobs = -1)
    grid.fit(features_mapped, target)
    print('Лучшие параметры:', grid.best_params_)
    # -------------------------------------------------------
    # Обучение модели
    best_model = RandomForestRegressor(n_estimators = grid.best_params_['n_estimators'],
                                      max_depth = grid.best_params_['max_depth'])
    best_model.fit(features_mapped, target)
    # -------------------------------------------------------
    # Моделирование и метрики
    predictions = best_model.predict(features_mapped)
    sMAPE = sMAPE_calc(target, predictions)
    print('sMAPE модели регрессии на обучающей выборке:', sMAPE)
    print("Коэффициент детерминации на обучающей выборке:", r2_score(target, predictions))
    # -------------------------------------------------------
    # Строим график
    plt.scatter(target, predictions, color='black')
    plt.xlabel('Target')
    plt.ylabel('Predictions')
    plt.title('График Target-Predictions на обучающей выборке')
    plt.show()
    return best_model, scaler_features
```

```python
def best_params_DecisionTreeRegressor(features, target):
    # -------------------------------------------------------
    # Масштабирование
    scaler_features = StandardScaler()
    scaler_features.fit(features)
    features_mapped = scaler_features.transform(features)
    # -------------------------------------------------------
    # Тип модели и её параметры
    model = DecisionTreeRegressor()
    parameters = {'max_depth': range(1, 10)}
    # -------------------------------------------------------
    sMAPE_s = make_scorer(sMAPE_calc, greater_is_better = False)
    grid = GridSearchCV(model, parameters, scoring = sMAPE_s, cv = 5, n_jobs = -1)
    grid.fit(features_mapped, target)
    print('Лучшие параметры:', grid.best_params_)
    # -------------------------------------------------------
    # Обучение модели
    best_model = DecisionTreeRegressor(max_depth = grid.best_params_['max_depth'])
    best_model.fit(features_mapped, target)
    # -------------------------------------------------------
    # Моделирование и метрики
    predictions = best_model.predict(features_mapped)
    sMAPE = sMAPE_calc(target, predictions)
    print('sMAPE модели регрессии на обучающей выборке:', sMAPE)
    print("Коэффициент детерминации на обучающей выборке:", r2_score(target, predictions))
    # -------------------------------------------------------
    # Строим график
    plt.scatter(target, predictions, color = 'black')
    plt.xlabel('Target')
    plt.ylabel('Predictions')
    plt.title('График Target-Predictions на обучающей выборке')
    plt.show()
    return best_model, scaler_features
```

#### CatBoostRegressor

```python
def best_params_CatBoostRegressor(features, target):
    # -------------------------------------------------------
    # Масштабирование
    scaler_features = StandardScaler()
    scaler_features.fit(features)
    features_mapped = scaler_features.transform(features)
    # -------------------------------------------------------
    # Тип модели и её параметры
    model = CatBoostRegressor()
    parameters = {'iterations': range(990, 1000, 10)}
                 
    # -------------------------------------------------------
    sMAPE_s = make_scorer(sMAPE_calc, greater_is_better = False)
    grid = GridSearchCV(model, parameters, scoring = sMAPE_s, cv = 5, n_jobs = -1)
    grid.fit(features_mapped, target)
    print('Лучшие параметры:', grid.best_params_)
    # -------------------------------------------------------
    # Обучение модели
    best_model = CatBoostRegressor(iterations = grid.best_params_['iterations'], silent = True)
    best_model.fit(features_mapped, target)
    # -------------------------------------------------------
    # Моделирование и метрики
    predictions = best_model.predict(features_mapped)
    sMAPE = sMAPE_calc(target, predictions)
    print('sMAPE модели регрессии на обучающей выборке:', sMAPE)
    print("Коэффициент детерминации на обучающей выборке:", r2_score(target, predictions))
    # -------------------------------------------------------
    # Строим график
    plt.scatter(target, predictions, color = 'black')
    plt.xlabel('Target')
    plt.ylabel('Predictions')
    plt.title('График Target-Predictions на обучающей выборке')
    plt.show()
    return best_model, scaler_features
```

Концентрации металлов на каждом этапе обработки различны и отображена их динамика, в т.ч. определены нижние выбросы для суммарной концентрации. Выбросы имеют причину сбой измерительного оборудования. Записи, содержащие в себе выбросы - удалены из всех множеств.

Распределения гранул сырья исследовано и выявлены их сходство для обучающей и тестовой выборок.


Обучены несколько моделей регрессии и определена модель, которая дает наилучший результат (CatBoost) для метрики sMAPE. Так же выполнено сравнение с наивной моделью DummyRegressor по медиане.


In this paper, the data of the technological process of a gold mining company were studied.


The concentrations of metals at each stage of processing are different and their dynamics are displayed, including the lower emissions for the total concentration are determined. The emissions are caused by a failure of the measuring equipment. Records containing outliers are removed from all sets.

The distribution of raw material granules was investigated and their similarity was revealed for the training and test samples.


Several regression models were trained and the model that gives the best result (Cut Boost) for the sMAPE metric was determined. A comparison with the naive DummyRegressor model by median is also performed.
