### The 'ceiling_height' column. Filling in gaps and correcting abnormal extremes.

```python
data.boxplot(column = 'ceiling_height')
plt.title('Boxplot для ceiling_height')
plt.show()
data['ceiling_height'].hist(bins = 50, range = (1, 5))
plt.title('Гистограмма значений ceiling_height на участке (1, 5)')
plt.show()
data['ceiling_height'].hist(bins = 50, range = (5, 23))
plt.title('Гистограмма значений ceiling_height на участке (5, 23)')
plt.show()
data['ceiling_height'].hist(bins = 50, range = (23, 40))
plt.title('Гистограмма значений ceiling_height на участке (23, 40)')
plt.show()
data['ceiling_height'].hist(bins = 50, range = (40, 100))
plt.title('Гистограмма значений ceiling_height на участке (40, 100)')
plt.show()
```

По строительным нормам потолки в квартирах строились и строятся высотой 2,4 - 4,5 м.

На графике boxplot видны явные выбросы максимумов.

Построим гистограммы для нескольких промежутков:

Промежуток (1, 5): видно, что распределение высоты потолков похоже на распределение Пуассона и распределение соответствует строительным нормам. Значения, которые меньше 2,4 м имеет смысл заменить медианным значением.

Промежуток (5, 23): включает в себя несколько значений и тоже их нужно заменить медианным значением.

Промежуток (23, 40): Здесь скопление значений. Вероятно, что это ошибочная высота потолков в дециметрах, а не в метрах и их нужно перевести в метры.

Промежуток (40, 100): Значения из этого промежутка будут заменены на медианное.

According to building standards, the ceilings in the apartments were built and are being built with a height of 2.4 - 4.5 m.

The boxplot graph shows clear outliers of the maxima.

Let's build histograms for several intervals:

Interval (1, 5): it can be seen that the ceiling height distribution is similar to the Poisson distribution and the distribution corresponds to building codes. It makes sense to replace values that are less than 2.4 m with the median value.

Interval (5, 23): includes several values and they also need to be replaced with the median value.

Interval (23, 40): Here is a cluster of values. It is likely that this is an erroneous ceiling height in decimeters, not in meters, and they need to be converted to meters.

Interval (40, 100): The values from this interval will be replaced by the median.

```python
ceiling_height_median = data['ceiling_height'].median()
print('Медиана ceiling_height =', ceiling_height_median)
min_ceiling = 2.4
max_ceiling = 4.5

def ceiling_correction(row):
    ceiling_height = row['ceiling_height']
    if ceiling_height < min_ceiling:
        return ceiling_height_median
    elif (ceiling_height > max_ceiling) & (ceiling_height < min_ceiling*10):
        return ceiling_height_median
    elif (ceiling_height >= min_ceiling*10) & (ceiling_height <= max_ceiling*10):
        return ceiling_height/10
    elif ceiling_height > max_ceiling*10:
        return ceiling_height_median
    else:
        return ceiling_height
    
data['ceiling_height'] = data.apply(ceiling_correction, axis = 1)
```

Расчитываем медиану после исправления аномальных значений и заполняем ею отсутствующие значения.

We calculate the median after correcting the abnormal values and fill in the missing values with it.

```python
ceiling_height_median_updated = data['ceiling_height'].median()
print('Медиана ceiling_height_updated =', ceiling_height_median_updated)
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median_updated)
print('Количество пропущенных значений в столбце ceiling_height:', data['ceiling_height'].isna().sum())
data['ceiling_height'].hist(bins = 10)
plt.title('Гистограмма значений ceiling_height итоговая')
plt.show()
```

```python
data.boxplot(column = 'floors_total')
plt.show()
display(data.query('floors_total > 50'))
```

Построен boxplot график для поиска выбросов. Два подозрительных здания с 52 и 60 этажей.

Изучение записей показывает:

- здание с 52 этажами существует в Санкт-Петербурге ("Балтийская Жемчужина")
- здание с 60 этажами в Кронштадте не существует и сделана опечатка 60 вместо 6 этажей.


A boxplot graph is built to search for outliers. Two suspicious buildings with 52 and 60 floors.

Studying the records shows:

- a building with 52 floors exists in St. Petersburg ("Baltic Pearl")
- a building with 60 floors does not exist in Kronstadt and a typo was made 60 instead of 6 floors.

```python
data['floors_total'] = data['floors_total'].replace(60, 6)
floors_total_median = data['floors_total'].median()
print('Медиана floors_total_median =', floors_total_median)
data['floors_total'] = data['floors_total'].fillna(floors_total_median)
print('Количество пропущенных значений в столбце floors_total:', data['floors_total'].isna().sum())
data['floors_total'].hist(bins = 10)
plt.title('Гистограмма значений floors_total')
plt.show()
data['floors_total'] = data['floors_total'].astype('int')
print('Этажность домов из списка:')
print(sorted(data['floors_total'].unique()))
```

Выполнена замена пропусков на медианное значение и сделано преобразование типа данных в целочисленный.

The omissions were replaced with the median value and the data type was converted to an integer.

```python
# Проверка floor > floors_total
display(data.query('floor > floors_total'))
print()
print('Количество записей с условием floor > floors_total:', len(data.query('floor > floors_total')))
# Замена floor на floors_total

def correction_floor_total(row):
    floor        = row['floor']
    floors_total = row['floors_total']
    if floors_total < floor:
        return floor
    else:
        return floors_total

data['floors_total'] = data.apply(correction_floor_total, axis = 1)
print('Количество записей с условием floor > floors_total после замены:', len(data.query('floor > floors_total')))
```

Выполнена замена этажности домов, для которых медианное значение оказалось меньше фактического этажа проживания. Каждый из этих этажей проживания теоретически может быть последним этажом в доме, т.к. дома с таким количеством этажей есть в списке уникальных значений. Реальная этажность неизвестна, поэтому она заменена этажом проживания.

The number of floors of houses for which the median value was less than the actual floor of residence was replaced. Each of these floors of residence can theoretically be the last floor in the house, since houses with so many floors are in the list of unique values. The actual number of floors is unknown, so it is replaced by the floor of residence.

```python
data.boxplot(column = 'living_area')
plt.title('Boxplot для living_area')
plt.show()
data['living_area'].hist(bins = 20, range = (0, 20))
plt.title('Гистограмма на интервале (0, 20) living_area')
plt.show()
data['living_area'].hist(bins = 5, range = (2, 2.5))
plt.title('Гистограмма на интервале (2, 2.5) living_area')
plt.show()
# Подфункция преобразования аномально низких значений
def living_area_small_num_correction(row):
    living_area = row['living_area']
    total_area = row['total_area']
    if (living_area < 2.5) & (total_area > living_area*10):
        return living_area*10
    else:
        return living_area
    
data['living_area'] = data.apply(living_area_small_num_correction, axis = 1)

# Определяем медиану площади одной комнаты, чтоб заполнить недостающие значения
room_area_arr = data['living_area']/data['rooms']
room_area_median = room_area_arr.median()
print('Медиана площади одной комнаты:', room_area_median)
print()
room_area_arr.hist(range = (1, 100))
plt.title('Гистограмма для room_area_arr')
plt.show()

def living_area_correction(row):
    living_area = row['living_area']
    rooms = row['rooms']
    if (living_area != living_area) & (rooms != 0):
        return room_area_median*rooms
    else:
        return living_area

data['living_area'] = data.apply(living_area_correction, axis = 1)
print('Количество NaN после первой очистки:', data['living_area'].isna().sum())

# Определяем медиану отношения living_area/total_area
living_ratio = data['living_area']/data['total_area']
living_ratio.hist()
plt.title('Гистограмма living_ratio')
plt.show()
living_ratio_median = living_ratio.median()
print('Медиана соотношения жилой и общей площади:', living_ratio_median)

def living_area_correction_2(row):
    living_area = row['living_area']
    total_area = row['total_area']
    if (living_area != living_area):
        return total_area*living_ratio_median
    else:
        return living_area
    
data['living_area'] = data.apply(living_area_correction_2, axis = 1)
print('Количество NaN после второй очистки:', data['living_area'].isna().sum())
```

Аномально низкие значения <2.5 м расценены, как ошибочные и умножены на 10.

Если известно значение ‘rooms’, то ‘living_area’ = ‘rooms’*room_area_median

Для оставшихся пропусков ‘living_area’ = ‘total_area’*’living_ratio_median’


The omissions were replaced in this column.

Abnormally low values <2.5 m are regarded as erroneous and multiplied by 10.

If the value 'rooms’ is known, then 'living_area' = 'rooms'*room_area_median

For the remaining omissions, 'living_area' = 'total_area'*'living_ratio_median’

#### Столбец 'last_price'. Устранение выбросов.

#### The 'last_price' column. Elimination of emissions.

Для устранения выбросов в стоимости квартиры нужно использовать стоимость одного квадратного метра. Установив параметры выбросов - выявить аномально высокие и аномально низкие значения.

To eliminate outliers in the cost of an apartment, you need to use the cost of one square meter. By setting the emission parameters, it is possible to identify abnormally high and abnormally low values.

```python
print('Статистика столбца sq_m_cost:')
display(data['sq_m_cost'].describe())
print()
print('Статистика столбца last_price:')
display(data['last_price'].describe())
print()
data['sq_m_cost'].hist(bins = 100)
plt.title('Гистограмма стоимости квадратного метра')
plt.show()
data.boxplot(column = 'sq_m_cost')
plt.title('Boxplot стоимости квадратного метра')
plt.show()
outlier_lower  = data['sq_m_cost'].quantile(0.25)-1.5*(data['sq_m_cost'].quantile(0.75)-data['sq_m_cost'].quantile(0.25))
print('Нижняя граница выброса: {:.1f}'.format(outlier_lower))
outlier_higher = data['sq_m_cost'].quantile(0.75)+1.5*(data['sq_m_cost'].quantile(0.75)-data['sq_m_cost'].quantile(0.25))
print('Верхняя граница выброса: {:.1f}'.format(outlier_higher))
print()

# Подфункция корректировки цены жилья
def last_price_correction(row):
    last_price = row['last_price']
    sq_m_cost  = row['sq_m_cost']
    total_area = row['total_area']
    if sq_m_cost > outlier_higher:
        return outlier_higher*total_area
    elif sq_m_cost < outlier_lower:
        return outlier_lower*total_area
    else:
        return last_price
    
data['last_price_corrected'] = data.apply(last_price_correction, axis = 1)
data['sq_m_cost_corrected'] = data['last_price_corrected']/data['total_area']
print('Выполнено замен:', (data['last_price'] != data['last_price_corrected']).sum())
print()
display('Статистика столбца last_price_corrected:', data['last_price_corrected'].describe())
print()
data['last_price_corrected'].hist(bins = 100)
plt.title('Гистограмма last_price после устранения выбросов')
plt.show()
print()
data['last_price_corrected'].hist(bins = 100, range = (0, 1500000))
plt.title('Гистограмма last_price_corrected после устранения выбросов на промежутке (0, 1500000)')
plt.show()
```

В столбце 'sq_m_cost' выявлены выбросы и их значения заменены верхней и нижней границами.

Добавлен столбец 'last_price_corrected'.

Причинами выбросов могут являться: ошибки ввода данных, мошеннические и договорные сделки.

Outliers were detected in the 'sq_m_cost' column and their values were replaced by upper and lower bounds.

Added the 'last_price_corrected' column.

The reasons for the outliers may be: data entry errors, fraudulent and contractual transactions.
